```python
for i in order_df['Region Name'].unique():
    forecast_group='Pr Plk'
    region=i
    df=order_df[order_df['Region Name']==i].reset_index(drop=True)
    # order_df['Order Create Date']=pd.to_datetime(order_df['Order Create Date'])
    df=df.groupby(by=['year','month','Region Name'])['Order Volume (STD)','Project Count'].sum().reset_index()


    df['date']=df['year'].astype('string')+'-'+df['month'].astype('string')+'-01'
    df['date']=pd.to_datetime(df['date'])
    df.set_index('date',inplace=True)
    df=df.drop(['year','month'],axis=1)
    print(df[['Order Volume (STD)','Project Count']].corr())
    ax=df[['Order Volume (STD)','Project Count']].apply(lambda x:(x-x.min())/(x.max()-x.min())).plot(figsize=(8,5),title=f'correlation for  {forecast_group} {region}')
    ax.axvline(x='2020-03-01',color='red', alpha = 0.6,linestyle='dashed')
    ax.axvline(x='2022-05-01',color='red', alpha = 0.6,linestyle='dashed')
    # ax.set_xticklabels(ax.get_xticklabels(), rotation=90)  # Adjust the rotation angle as needed
    plt.show()
```
